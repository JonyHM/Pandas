```python
%matplotlib inline
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
obes_02 = pd.read_csv('g07_02.csv', sep=';', encoding='ISO-8859-1', low_memory=False)

obes_08 = pd.read_csv('g07_08.csv', sep=';', encoding='ISO-8859-1', low_memory=False)

#Lendo os dois arquivos CSV disponíveis nos dados pesquisados
```

```python
obes_02 = obes_02.drop('Sobrepeso', axis=1) #Remove a coluna 'Sobrepeso' do aquivo

for c in range(1,15):
    obes_02 = obes_02.drop('Sobrepeso.'+str(c), axis=1)
        
obes_02

#Remove as demais colunas 'Sobrepeso' numeradas de 1 a 14 e as exibe
```

```python
obes_08 = obes_08.drop('Sobrepeso', axis=1) #Remove a coluna 'Sobrepeso' do aquivo

for c in range(1,15):
    obes_08 = obes_08.drop('Sobrepeso.'+str(c), axis=1)

obes_08
```

```python
obes_02 = obes_02.drop('Excesso de peso', axis=1) #Remove a coluna 'Excesso de peso' do aquivo

for c in range(1,15):
    obes_02 = obes_02.drop('Excesso de peso.'+str(c), axis=1)
obes_02

#Remove as demais colunas 'Excesso de peso' numeradas de 1 a 14 e as exibe
```

```python
obes_08 = obes_08.drop('Excesso de peso', axis=1) #Remove a coluna 'Excesso de peso' do aquivo

for c in range(1,15):
    obes_08 = obes_08.drop('Excesso de peso.'+str(c), axis=1)
obes_08

#Remove as demais colunas 'Excesso de peso' numeradas de 1 a 14 e as exibe
```

```python
obes_02.dropna(axis=0, how='all', inplace=True) #Remove as linhas sem dados
```

```python
obes_08.dropna(axis=0, how='all', inplace=True) #Remove as linhas sem dados
```

```python
obes_02
```

```python
obes_08
```

```python
obes_02 = obes_02.drop([11]) #remove a linha com índice 11 do arquivo
```

```python
obes_08 = obes_08.drop([11,12,13]) #remove as linhas com índices 11, 12, 13 do arquivo
```

```python
obes_02.rename(columns = {"Unnamed: 0":""}, inplace=True) #Renomeando a coluna com nome "Unnamed: 0" para ""
```

```python
obes_08.rename(columns = {"Unnamed: 0":""}, inplace=True) #Renomeando a coluna com nome "Unnamed: 0" para ""
```

```python
obes_02
```

```python
obes_08
```

```python
# Retirando colunas indesejáveis, onde estão dados separados por gênero, para que sobre apenas os dados totais

indice = 1
manter = 2
obes_02_total = obes_02

obes_02_total = obes_02_total.drop('Obesidade', axis=1)

while indice < 15:
    if indice == manter:
        obes_02_total = obes_02_total
        manter += 3
    else:
        obes_02_total = obes_02_total.drop('Obesidade.'+str(indice), axis=1)
        
    indice += 1
```

```python
obes_02_total
```

```python
# Retirando colunas indesejáveis, onde estão dados separados por gênero, para que sobre apenas os dados totais

indice = 1
manter = 2
obes_08_total = obes_08

obes_08_total = obes_08_total.drop('Obesidade', axis=1)

while indice < 15:
    if indice == manter:
        obes_08_total = obes_08_total
        manter += 3
    else:
        obes_08_total = obes_08_total.drop('Obesidade.'+str(indice), axis=1)
        
    indice += 1
```

```python
obes_08_total
```

```python
# Renomeando colunas com nomes semelhantes, para o nome abaixo delas na tabela anterior

obes_02_total.rename(columns = {"":"Região", "Obesidade.2":"20 a 29 anos", "Obesidade.5":"30 a 39 anos", "Obesidade.8":"40 a 49 anos", "Obesidade.11":"50 a 59 anos", "Obesidade.14":"Total"}, inplace=True)
obes_08_total.rename(columns = {"":"Região", "Obesidade.2":"20 a 29 anos", "Obesidade.5":"30 a 39 anos", "Obesidade.8":"40 a 49 anos", "Obesidade.11":"50 a 59 anos", "Obesidade.14":"Total"}, inplace=True)
```

```python
# Removendo as colunas não mais utilizadas
obes_02_total = obes_02_total.drop([0,1])
obes_08_total = obes_08_total.drop([0,1])
```

```python
# Renomeando a coluna 'Total', pois tem nome semelhante a uma linha (índice 7)

obes_02_total.rename(columns = {"Total":"Total_idade"}, inplace=True)
obes_08_total.rename(columns = {"Total":"Total_idade"}, inplace=True)
```

```python
obes_02_total
```

```python
obes_08_total
```

```python
# Criando um Data Frame com os dados das idades em geral

obes_02_total_final = pd.DataFrame(obes_02_total, columns=['Região', 'Total_idade'])
```

```python
# Organizando os dados da coluna "Total_idade" em ordem crescente

obes_02_total_final = obes_02_total_final.sort_values(by='Total_idade')
obes_02_total_final
```

```python
# Como os dados foram organizados com base no primeiro algarismo dos números na coluna "Total_idade", 
#eles foram renomeados com um '0' antes para que seja organizado corretamente

obes_02_total_final.iat[3, 1] = '08,9'
obes_02_total_final.iat[4, 1] = '09,2'
obes_02_total_final.iat[5, 1] = '09,5'
```

```python
# Após organizar novamente, a ordem fica correta
obes_02_total_final = obes_02_total_final.sort_values(by='Total_idade')
obes_02_total_final
```

```python
# Configurando a posição e tamanho das barras
pos = list(range(len(obes_02_total_final['Total_idade']))) 
width = 0.50 

# Plotando as barras
fig, ax = plt.subplots(figsize=(10,5))

# Criando uma barra com os dados de 'Total',
# na posição 'pos',
plt.bar([p + width for p in pos], 
        #usando dados de 'obes_02_total_final['Total_idade']',
        obes_02_total_final['Total_idade'], 
        # com tamanho:
        width, 
        # espessura 0.7:
        alpha=0.7, 
        # cor:
        color='#087833', 
        # e com o primeiro valor em 'Total_idade'] como rótulo
        label=obes_02_total_final['Total_idade']) 

# Rótulo de y
ax.set_ylabel('Porcentagem')

# Título da tabela
ax.set_title('Obesidade no Brasil')

# posição da legenda de x
ax.set_xticks([p + width for p in pos])

# leganda de x
ax.set_xticklabels(obes_02_total_final['Região'])

# adicionando a legenda da tabela e plotando o gráfico
plt.legend(['Brasil'], loc='upper left')
plt.grid()
plt.show()
```
